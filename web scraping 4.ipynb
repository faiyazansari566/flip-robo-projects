```python
##name
try :
    elements = driver.find_elements_by_xpath("//table[@cellpadding='3']/tbody/tr/td[2]/table[2]/tbody/tr/td/table/tbody/tr/td[2]")
    for i in elements :
        if i.text is None :
            name.append("-")
        else :
            name.append(i.text)
except NoSuchElementException:
    name.append("-")
    
##data type
try :
    elements = driver.find_elements_by_xpath("//table[@cellpadding='3']/tbody/tr/td[2]/table[2]/tbody/tr/td[2]")
    for i in elements :
        if i.text is None :
            dtype.append("-")
        else :
            dtype.append(i.text)
except NoSuchElementException:
    dtype.append("-")

##task
try :
    elements = driver.find_elements_by_xpath("//table[@cellpadding='3']/tbody/tr/td[2]/table[2]/tbody/tr/td[3]")
    for i in elements :
        if i.text is None :
            task.append("-")
        else :
            task.append(i.text)
except NoSuchElementException:
    task.append("-")
 

##attribution type
try :
    elements = driver.find_elements_by_xpath("//table[@cellpadding='3']/tbody/tr/td[2]/table[2]/tbody/tr/td[4]")
    for i in elements :
        if i.text is None :
            attribute_type.append("-")
        else :
            attribute_type.append(i.text)
except NoSuchElementException:
    attribute_type.append("-")
    
##no_of_instances
try :
    elements = driver.find_elements_by_xpath("//table[@cellpadding='3']/tbody/tr/td[2]/table[2]/tbody/tr/td[5]")
    for i in elements :
        if i.text is None :
            no_of_instances.append("-")
        else :
            no_of_instances.append(i.text)
except NoSuchElementException:
    no_of_instances.append("-")
    
##no_of_attributes
try :
    elements = driver.find_elements_by_xpath("//table[@cellpadding='3']/tbody/tr/td[2]/table[2]/tbody/tr/td[6]")
    for i in elements :
        if i.text is None :
            no_of_attributes.append("-")
        else :
            no_of_attributes.append(i.text)
except NoSuchElementException:
    no_of_attributes.append("-")
    

##year
try :
    elements = driver.find_elements_by_xpath("//table[@cellpadding='3']/tbody/tr/td[2]/table[2]/tbody/tr/td[7]")
    for i in elements :
        if i.text is None :
            year.append("-")
        else :
            year.append(i.text)
except NoSuchElementException:
    year.append("-")
```
